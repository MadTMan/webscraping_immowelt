```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.immowelt.de/classified-search?distributionTypes=Buy&estateTypes=House&locations=AD08DE2112&m=homepage_new_search_classified_search_result'

# Send a GET request to the website
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.text, 'html')
print(soup)
```

```python
# Find the data containers (e.g., property listings)
properties = soup.find_all('div', class_='css-79elbk')
properties
```

```python
property_list = []
for prop in properties:
    title = prop.find('a', class_='css-xt08q3')['title']
    about = prop.find('div', class_='css-1cbj9xw').get_text()
    location = prop.find('div', class_ = 'css-162g046').get_text()
    
#     size_tag = prop.find('div', class_ = 'css-9u48bm').get_text()
#     rroom = size_tag[0]
#     sizee = size_tag[1]

    
    div_tag = soup.find('div', {'data-testid': 'classified-card-mfe-24VE6JPGN35F'})
    # Find the <a> tag within the div tag
    a_tag = div_tag.find('a')
    # Extract the href attribute
    link = a_tag['href']
    
    details = title.split(' - ')
    property_type = details[0]
    price = details[2]
    rooms = details[3]

    print(f"Property Type: {property_type}")
    print(f"About: {about}")
    print(f"Location: {location}")
    print(f"Price: {price}")
    print(f"Rooms: {size}")
    print(f"Size: {rooms}")
    print(f"link: {link}")
    property_list.append({'Title': title,'About' : about,'Price': price, 'Location': location, 'Rooms': size, 'Size': rooms, })


# Detached house for sale - Düsseldorf - 435,900 € - 5 rooms, 183 m²
```

```python

# Convert the list to a DataFrame
df = pd.DataFrame(property_list)

# # Save the DataFrame to a CSV file
# df.to_csv('properties.csv', index=False)

df
```
